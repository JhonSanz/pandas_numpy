### Making changes on the data.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('pokemon_data.csv')
df
```

```python
# Adding new column with computed values.
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df[0:5]
```

```python
# And then delete it.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
df = df.drop(columns=['Total'])
df.head(5)
```

##### For the column sum example check this beautiful way to do that.

```python
# First, we are selecting all dataframe rows from 4th to 10th column 
# Then, axis=1 means sum horizontally, 0 is vertically.
# So for every row we sum the values selected by .iloc() and then put it into Total column.
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df
```

```python
# Here we can see that .sum() function works over Dataframes.
# This example is an unziped version of the previous example.
# Important is to know that .sum() can compute for every row by itself.
df['Total2'] = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum(axis=1)
df
```

```python
# And probably exists many ways to do that...
df.eval('TotalN = HP + Attack + Defense + `Sp. Atk` + `Sp. Def` + Speed')
```

### Modifying an specific value

```python
df2 = df.drop(columns=['Total', 'Total2']).head(5)
df2
```

```python
# There are many ways to do that, let's check those
df2.at[0, 'Name'] = 'Vulvasaur'
df2
```

```python
df2.iloc[2, 2:4] = 'test'
df2
```

```python
df2.loc[3, ['Attack']] = 111
df2
```

```python
df2[['Defense']] = 777
df2
```

```python
# You also can reorder the columns. I don't know why but you can do that.
columns = list(df.columns.values)
df2 = df[columns[0:2] + [columns[-1]] + columns[2:-1]]
df2
```

### Conditional changes

```python
# Notice that you can use .loc like this
df.loc[[2, 3], ['Generation']] = 777
df
# Here we selected the rows as first argument, and column in the second place
# and then modify the value
```

```python
# So you can SELECT THE ROWS WITH A CONDITION
df['Legendary'] = False
df.loc[df['Type 1'] == 'Fire', ['Generation', 'Legendary']] = [99, True]
df
# Only Type 1 == fire will be Legendaries
# Notice that you can modify both values at the same time
```
