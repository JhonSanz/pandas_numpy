### Indexing

````python
import pandas as pd
df = pd.read_csv('admission_predict.csv')
df
````

````python
# Now we want to have our Chance of Admit column as index
df['Serial number'] = df.index
df = df.set_index('Chance of Admit ') # Here is useful using .strip
df
````

````python
# We can use reset index to generate a new column with the original serial Numbers
# and it's going to promote our Chance of Admit to a column
df = df.reset_index()
df
````

````python
# Similary we can keep specific columns like this
columns_to_keep = [
    'STNAME', 'CTYNAME',
    'BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012', 'BIRTHS2013',
    'BIRTHS2014', 'BIRTHS2015',
    'POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012',
    'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015'
]
df = df[columns_to_keep]
df.head()
````

Here we are going to set a _multi index_. Notice that we have **several cities in the same state.** For example, we can think it like this:

```python
{
    "Alabama": [
        "Alabama", "Autauga County", "Baldwin County",
        "Barbour Count"," Bibb County", "..."
    ],
    "...": ["..."],
    "Wyoming": [
        "Sweetwater County", "Teton County",
        "Uinta County", "..."
    ],
}

```

So, let's do that for our data.

````python
df = df.set_index(['STNAME', 'CTYNAME'])
df
````

````python
# To query this dataframe now we need to provide the arguments in order by the level we whish to query.
# We want to see the population results from Washtenaw County in Michigan the state
df.loc['Michigan', 'Washtenaw County']
````

````python
# Here we got a warning explained here
# https://stackoverflow.com/questions/54307300/what-causes-indexing-past-lexsort-depth-warning-in-pandas
# To fix it we have to use .sort_index
df = df.sort_index()
df.loc['Michigan', 'Washtenaw County']
````

````python
# Also we can use .loc tipicaly but passing a tuples for both indexes as first argument
# and in this example we are selecting only two columns as usual
df.loc[
    [
        ('Michigan', 'Washtenaw County'),
        ('Michigan', 'Wayne County')
    ],
    ['BIRTHS2010', 'BIRTHS2011']
]
````
