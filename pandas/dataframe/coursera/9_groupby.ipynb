### Group by

The idea behind the groupby() function is that it takes some dataframe, splits it into chunks based on some key values, applies computation on those chunks, then combines the results back together into another dataframe. In pandas this is refered to as the split-apply-combine pattern.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('census.csv')
df = df[df["SUMLEV"] == 50]
df.head()
```

- Let's get a list of the unique states,
- then we can iterate over all the states and for each state we reduce the data frame
- and calculate the average.

```python
print(df["STNAME"].unique())
df[df["STNAME"] == 'Alabama']
```

```python
print(np.average(df[df["STNAME"] == 'Alabama']["CENSUS2010POP"]))
df[df["STNAME"] == 'Alabama']["CENSUS2010POP"]
```

```python
# Now do that for every estates
for state in df["STNAME"].unique():
    avg = np.average(df.where(df["STNAME"] == state).dropna()["CENSUS2010POP"])
    # avg = np.average(df[df["STNAME"] == state]["CENSUS2010POP"]) This version is faster!
    print(
        f"Counties in state {state}" +
        f" have an average population of {avg}"
    )
```

Now do that with group by

```python
to_print = None
for group, frame in df.groupby('STNAME'):
    # You'll notice there are two values we set here. groupby() returns a tuple, where the first value is the
    # value of the key we were trying to group by, in this case a specific state name, and the second one is
    # projected dataframe that was found for that group
    print(group)
    to_print = frame.copy()
    break
print(to_print["STNAME"].unique())
to_print
# Here we can see that frame is a dataframe for Alabama
# We avoided to query the dataframe, and just grouped it by STNAME
```

```python
for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print(
        f"Counties in state {state}" +
        f" have an average population of {avg}"
    )
# It's Faster!
```

Now we are going to use groupby with a function. We wnat to divide our data within groups that matches these intervals:

- A-M
- M-Q
- Q-Z

```python
df = df.set_index("STNAME")
def set_batch_number(item):
    if item[0] < 'M':
        return 0
    if item[0] < 'Q':
        return 1
    return 2

for group, frame in df.groupby(set_batch_number):
    print(f"There are {len(frame)} records in group {group} for processing")
```

```python
df = pd.read_csv('listings.csv')
df.head()
```

```python
# So, how would I group by both of these columns? A first approach might be to promote them to a multiindex
# and just call groupby()
df = df.set_index(["cancellation_policy", "review_scores_value"])
df.head()
```

```python
to_print = None
for group, frame in df.groupby(level=0):
    print(group)
    to_print = frame.copy()
    break
to_print.head(2)
```

```python
# When we have a multiindex we need to pass in the leves that we want to group
to_print = None
for group, frame in df.groupby(level=1):
    print(group)
    to_print = frame.copy()
    break
to_print.head(2)
```

```python
# When we have a multiindex we need to pass in the leves that we want to group
i = 0
to_print = None
for group, frame in df.groupby(level=(0, 1)):
    print(group)
    if i == 0:
        to_print = frame.copy()
    else:
        to_print = pd.concat([to_print, frame.copy()])
    i += 1
    if i == 5:
        break
print(i)
to_print

# As you can see, for every iteration we can get a frame with multiple records
# but the group are just the rows that we defined as indexes
```

Now check this example. Here we are going to separete our data in two groups and notice that the iterations are going to decrease

```python
def grouping_funtion(item):
    if item[1] == 10.0:
        return (item[0], "10.0")
    return (item[0], "not 10.0")

for group, frame in df.groupby(by=grouping_funtion):
    print(group)
```

```python
i = 0
to_print = None
for group, frame in df.groupby(by=grouping_funtion):
    print(group)
    to_print = frame.copy()
    if i == 1:
        break
    i += 1
to_print

# Here we are printing the group for "flexible" that doesn't have 10.0 as review_scores_value
```

### Aggregation

After grouping_by we need to do something with the data. Take a look

```python
df = pd.read_csv('listings.csv')
df.head()
```

```python
# With .agg we select the row that we are interested in
df.groupby('cancellation_policy').agg({'review_scores_value': np.average})
# Here we got just NaN values, but is because np.average doesn't ignore NaN values
```

```python
# With nanmean we ensure that if is there any NaN value on our frames
# it can compute the mean with no problems
df.groupby('cancellation_policy').agg({'review_scores_value': np.nanmean})
```

```python
# Now you can choose multiple columns with multiple funtions in the .agg method
df.groupby('cancellation_policy').agg({
    'review_scores_value': [np.nanmean, np.std],
    'reviews_per_month': np.nanmean
})
```

### Transformation

```python
cols = ['cancellation_policy', 'review_scores_value']
df[cols]
```

```python
to_print = None
for group, frame in df[cols].groupby('cancellation_policy'):
    print(group)
    to_print = frame.copy()
    break
to_print
```

```python
transform_df = df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.rename({'review_scores_value': 'transform_review_scores_value'}, axis='columns', inplace=True)
transform_df
```

```python
df = df.merge(transform_df, left_index=True, right_index=True)
df
```

Ĺo que pasó aqui es que tomé todo el grupo para calcular la media y luego se la asigné a cada registro por grupo en la tabla.

```python
# Now we can perform any operations with the data computed.
df['mean_diff'] = np.absolute(df['review_scores_value'] - df['transform_review_scores_value'])
df['mean_diff']
```

### Applying

Finally one of the most used functions after groupby. It allows to apply any arbitrary function

```python
df = pd.read_csv('listings.csv')
df = df[['cancellation_policy', 'review_scores_value']]
df
```

```python
# Let's do the same example: compute the mean
def calc_mean_review_scores(group):
    avg = np.nanmean(group['review_scores_value'])
    group['review_scores_mean'] = np.abs(avg - group['review_scores_value'])
    return group

df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```
