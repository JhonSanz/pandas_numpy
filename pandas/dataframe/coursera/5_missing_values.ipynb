### Missing Values

```python
import pandas as pd
import numpy as np
df = pd.read_csv('class_grades.csv')
df
```

```python
# We can use the function .isnull to create a mask appliying isnull() in every cell on the dataframe
# It can be useful for proccesing rows based on certaing column of data
mask = df.isnull()
mask
```

```python
# Here dropna takes all the rows that has NaN values and drop those rows
df.dropna()
```

```python
# Also exists fillna function to set values in NaN cells
df.fillna(666, inplace=True)
df
# Notice that inplace changes the values by reference
```

---

```python
# Now use parameter() with ffill and bfill
# Those are logs about user interaction with a video player.
df = pd.read_csv('log.csv')
df
```

```python
# We are going to sort and reindex those records.
df = df.set_index('time')
df = df.sort_index()
df
```

```python
# As we can see two users can interact at the same time. So the indexes are not unique
df = df.reset_index()
df
```

```python
# And now we can group our records by time.
df = df.set_index(['time', 'user'])
df
```

```python
# Now use fillna to fill our NaN values
df = df.fillna(method='ffill')
df
```

```python
df = pd.DataFrame({
    "A": [1, 1, np.NaN, 777, 4],
    "B": [np.NaN, 666, 4, 7, 6],
    "C": ["a", "b", "c", "d", "e"]
})
df
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='ffill')
```
